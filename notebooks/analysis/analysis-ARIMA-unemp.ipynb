## Heteroscedasticity

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

# Function to perform heteroscedasticity test
def heteroscedasticity_test(series):
    # Create the independent variable (X) and dependent variable (y)
    y = series.values
    X = np.arange(len(series)).reshape(-1, 1)  # Simple time index as a predictor
    X = sm.add_constant(X)  # Add intercept

    # Fit OLS model
    model = sm.OLS(y, X).fit()

    # Perform Breusch-Pagan test
    test_stat, p_value, _, _ = het_breuschpagan(model.resid, X)

    return p_value  # Return p-value of the test

# Apply test to each column
results = {col: heteroscedasticity_test(unemployment_pivot_diff_df["unemployment_rate"]) for col in unemployment_pivot_diff_df.columns}

# Display results
for col, p_value in results.items():
    print(f"Heteroscedasticity test for {col}: p-value = {p_value:.4f}")
```

## Normality

```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro, normaltest, anderson

# Function to apply normality tests
def normality_tests(series):
    results = {}
    
    # Shapiro-Wilk Test
    stat, p = shapiro(series)
    results['Shapiro-Wilk'] = {'statistic': stat, 'p-value': p}
    
    # D'Agostino and Pearson's Test
    stat, p = normaltest(series)
    results['D’Agostino-Pearson'] = {'statistic': stat, 'p-value': p}
    
    # Anderson-Darling Test
    anderson_result = anderson(series, dist='norm')
    results['Anderson-Darling'] = {
        'statistic': anderson_result.statistic,
        'critical_values': anderson_result.critical_values,
        'significance_levels': anderson_result.significance_level,
    }
    
    return results

# Apply tests to each column
normality_results = {col: normality_tests(unemployment_pivot_diff_df["unemployment_rate"]) for col in df.columns}

# Display results
for col, tests in normality_results.items():
    print(f"\nNormality tests for {col}:")
    for test, result in tests.items():
        print(f"  {test}: {result}")
```

## Kurtosis

```python
from scipy.stats import kurtosis
# Function to calculate kurtosis for a DataFrame
def calculate_kurtosis(df, fisher=False):
    """
    Calculate kurtosis for each column in the DataFrame.
    
    Args:
        df: A pivoted DataFrame where each column is a time series.
        fisher: Whether to return Fisher kurtosis (subtracting 3) or Pearson kurtosis.
        
    Returns:
        A dictionary with column names as keys and their kurtosis as values.
    """
    kurtosis_values = {col: kurtosis(df[col], fisher=fisher, nan_policy='omit') for col in df.columns}
    return kurtosis_values

# Calculate kurtosis
kurtosis_results = calculate_kurtosis(unemployment_pivot_df, fisher=False)

# Display results
for series, kurt in kurtosis_results.items():
    print(f"Kurtosis for {series}: {kurt:.2f}")
```

# ARIMA

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model
AR_2_1_0 = ARIMA(unemployment_pivot_df['unemployment_rate'], order=(2, 0, 0)).fit() # Specify (p, d, q)

# Print model summary
print(f"{AR_2_1_0.summary()}\n")
```

```python
import numpy as np

# Model
model = AR_2_1_0

# Data
data = unemployment_pivot_df

# Parameters
phi_1 = model.arparams[0]
phi_2 = model.arparams[1]
sigma = np.sqrt(model.params['sigma2'])
mu = np.mean(data['unemployment_rate'])
forecast_horizon = 30
num_scenarios = 73

# Simulate AR(2) process
scenarios = np.zeros((forecast_horizon, num_scenarios))
initial_values = data['unemployment_rate'].iloc[-2:]  # Last tthree observations

for i in range(num_scenarios):
    X_t_1, X_t_2 = initial_values[-1], initial_values[-2]
    scenarios[0, i] = X_t_1
    for t in range(1, forecast_horizon):
        epsilon_t = np.random.normal(0, sigma)
        X_t = mu + phi_1 * (X_t_1 - mu) + phi_2 * (X_t_2 - mu) + epsilon_t
        scenarios[t, i] = X_t
        X_t_2, X_t_1 = X_t_1, X_t

print(f"AR(2) Coefficients: phi_1={phi_1}, phi_2={phi_2}, sigma^2={sigma}")
```

```python
import matplotlib.pyplot as plt

# Plot the AR(3) simulated scenarios
def plot_scenarios(scenarios, forecast_horizon, num_scenarios, initial_values, mu):
    plt.figure(figsize=(12, 6))

    # Plot each scenario
    for i in range(num_scenarios):
        plt.plot(range(forecast_horizon), scenarios[:, i], color='green', alpha=0.2)

    # Plot the long-term mean (mu)
    plt.axhline(y=mu, color='blue', linestyle='--', label='Long-Term Mean (μ)')

    # Plot initial values
    plt.scatter([-1, 0], initial_values, color='red', label='Initial Values')

    # Add labels and title
    plt.title(f'Simulated AR(3) Scenarios ({num_scenarios} Paths)')
    plt.xlabel('Time Steps')
    plt.ylabel('Unemployment Rate')
    plt.legend()
    plt.grid(alpha=0.5)
    plt.show()

# Call the plot function
plot_scenarios(
    scenarios=scenarios,
    forecast_horizon=forecast_horizon,
    num_scenarios=num_scenarios,
    initial_values=initial_values,
    mu=mu
)
```
