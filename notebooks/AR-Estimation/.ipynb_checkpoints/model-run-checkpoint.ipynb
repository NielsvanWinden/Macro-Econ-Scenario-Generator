```python
import numpy as np
import pandas as pd

# Assign random seed
np.random.seed(42)

# Parameters
n_simulations = 1000       # Number of Monte Carlo simulations
n_steps = 500              # Time steps
dt = 0.01                  # Time increment
n_variables = 3            # Number of interrelated variables
lags = [1, 2]              # Lag values in time steps

# OU parameters
theta = np.array([0.5, 0.4, 0.6])  # Mean reversion speeds
mu = np.array([0.0, 0.0, 0.0])     # Long-term means
sigma = np.array([0.2, 0.2, 0.2])  # Volatility

# Lagged correlation matrices for each lag
phi_lag = {
    1: np.array([[0.0, 0.2, 0.1],  # Correlation matrix for lag = 1
                 [0.2, 0.0, 0.3],
                 [0.1, 0.3, 0.0]]),
    2: np.array([[0.0, 0.1, 0.2],  # Correlation matrix for lag = 2
                 [0.1, 0.0, 0.4],
                 [0.2, 0.4, 0.0]])
}

# Initialize storage
simulations = np.zeros((n_simulations, n_steps, n_variables))

# Run simulations
for sim in range(n_simulations):
    x = np.zeros((n_steps, n_variables))  # Initialize variables
    for t in range(1, n_steps):
        # Mean reversion term
        mean_reversion = theta * (mu - x[t-1])
        # Random noise term
        noise = sigma * np.random.normal(size=n_variables)
        # Lagged influence term
        lag_effect = np.zeros(n_variables)
        for lag, phi in phi_lag.items():  # Iterate over lags and their correlation matrices
            if t - lag >= 0:  # Check if the lag is valid
                for i in range(n_variables):
                    for j in range(n_variables):
                        if i != j:
                            lag_effect[i] += phi[i, j] * x[t-lag, j]
        # Update process
        x[t] = x[t-1] + (mean_reversion + lag_effect) * dt + noise * np.sqrt(dt)
    simulations[sim] = x

# Convert to DataFrame for easier analysis
time = np.arange(n_steps) * dt
simulated_data = [pd.DataFrame(simulations[sim], columns=[f"Variable_{i+1}" for i in range(n_variables)], index=time)
                  for sim in range(n_simulations)]

# Example: Display one simulation
print(simulated_data[0].head())
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 2  # Number of components
T = 100  # Number of time steps
dt = 1  # Time step size
A = np.array(
    [[0.5, 0.1], # Drift matrix
    [0.2, 0.4]]
)  
Sigma = np.array(
    [[0.2, 0.05], # Noise covariance matrix        
     [0.05, 0.3]]
)  

# Externally provided baseline mu_t (time-varying)
mu = np.zeros((T, n))
mu[:, 0] = np.linspace(1, 10, T)  # Example: Linear increase for first component
mu[:, 1] = np.linspace(5, -5, T)  # Example: Linear decrease for second component

# Initialize the process
X = np.zeros((T, n))
noise = np.random.multivariate_normal(mean=np.zeros(n), cov=Sigma, size=T)

# Simulate the process
for t in range(T - 1):
    drift = A @ (mu[t] - X[t]) * dt
    diffusion = noise[t] * np.sqrt(dt)
    X[t + 1] = X[t] + drift + diffusion

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(X[:, 0], label="X1 (Simulated)")
plt.plot(mu[:, 0], linestyle='--', label="mu1 (Baseline)")
plt.plot(X[:, 1], label="X2 (Simulated)")
plt.plot(mu[:, 1], linestyle='--', label="mu2 (Baseline)")
plt.legend()
plt.title("SDE Simulation with Externally Provided Baseline Mu")
plt.show()
```
